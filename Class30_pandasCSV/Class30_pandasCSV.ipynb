# Convert DataFrame to CSV or xlsx

```python
dict = {'Name':pd.Series(['Tom', 'Jack', 'Steve', 'Ricky', 'Vin', 'James', 'Smith']),
       'Age':pd.Series([25,26,25,35,23,33,31]),
       'Rating':pd.Series([4.23,4.1,3.4,5,2.9,4.7,3.1])}

df = pd.DataFrame(dict)
df
```

```python
df.to_csv('first.csv')
```

```python
df.to_csv('first.csv', index=False)
```

```python
df.to_excel('first.xlsx',sheet_name='Emp',index=False)
```
