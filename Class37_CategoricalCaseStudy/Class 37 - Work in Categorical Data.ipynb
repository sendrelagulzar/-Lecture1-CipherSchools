```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 3 Working With Categorical Data

```python
tips = pd.read_csv('tips.csv')

tips.head()
```

```python
tips.shape
```

```python
tips.describe(include='all')
```

```python
tips['day'].value_counts()
```

### 3.1 Categorical Scatterplots - stripplot()

bivariate

```python
sns.stripplot(data = tips, x = 'day', y='total_bill', hue='day')
```

### 3.2 Swarm Plot

This function is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, but it does not scale well to large numbers of observations. This style of plot is sometimes called a “beeswarm”.

```python
sns.swarmplot(data = tips, x='day', y='total_bill')
```

### 3.3 Boxplot

```python
sns.boxplot(data = tips, x='day', y='total_bill')
```

```python
tips.loc[tips.time == 'Lunch', :]
```

```python
tips.loc[tips.time == 'Lunch', 'day'].value_counts()
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='time')
```

```python
# Adding a new column, indicating weekend or not

tips['weekend'] = tips['day'].isin(['Sat', 'Sun'])

tips.head()
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='weekend')
```

### 3.4 Bar Plots

shows average

```python
titanic = pd.read_csv('titanic.csv')

titanic.head()
```

```python
titanic.shape
```

```python
titanic.describe(include='all')
```

```python
sns.barplot(x='sex', y='age', data=titanic)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(121)
sns.barplot(x='sex', y='age', data=titanic)
plt.title('Avg Age')

plt.subplot(122)
sns.barplot(x='sex', y='age', data=titanic, estimator = np.median)
plt.title('Median Age')

plt.show()
```

```python
# Replace missing values in the 'age' column with a default value (e.g., 0)
titanic['age'].fillna(np.mean, inplace=True)
```

```python
plt.figure()

plt.subplot(1,2,1)
sns.barplot(x='sex', y='age', data=titanic)
plt.title('Avg Age')

plt.subplot(1,2,2)
sns.barplot(x='sex', y='age', data=titanic, estimator = np.median)
plt.title('Median Age')

plt.show()
```

```python
sns.boxplot(x='sex', y='age', data=titanic)
```

```python
sns.barplot(x='sex', y='survived', data=titanic)
```

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

# Case Study: Covid 19

### using time series analysis

time series analysis is done by converting object to date and time, the using lineplot

```python
ca_df = pd.read_csv('covid-19/countries-aggregated.csv')

ww_df = pd.read_csv('covid-19/worldwide-aggregated.csv')
```

```python
print(ca_df.shape)

ca_df.head()
```

```python
print(ww_df.shape)

ww_df.head()
```

```python
ca_df.info()
```

```python
ww_df.info()
```

```python
ww_df['Date'] = pd.to_datetime(ww_df['Date'])

ww_df.info()
```

```python
plt.figure(figsize=(16, 8))

sns.lineplot(x = 'Date', y = 'Confirmed', data = ww_df)

plt.xticks(rotation = 45)

plt.show()
```

```python
ca_df['Date'] = pd.to_datetime(ca_df['Date'])

ca_df.info()
```

```python
group_ca_df = ca_df.groupby('Country')

group_ca_df.first()
```

```python
group_ca_df.groups.keys()
```

```python
plt.figure(figsize=(18, 14))

plt.subplots_adjust(hspace=0.5)

plt.subplot(311)
sns.lineplot(x = 'Date', y = 'Confirmed', data = group_ca_df.get_group('India'))
plt.title('India')

plt.subplot(312)
sns.lineplot(x = 'Date', y = 'Confirmed', data = group_ca_df.get_group('China'))
plt.title('China')

plt.subplot(313)
sns.lineplot(x = 'Date', y = 'Confirmed', data = group_ca_df.get_group('US'))
plt.title('US')

plt.show()
```

```python

temp_df = ca_df.loc[(ca_df.Country=='China') | (ca_df.Country=='India') | (ca_df.Country=='US'), : ]

temp_df.head()
```

```python
temp_df.info()
```

```python
plt.figure(figsize=(16, 8))

sns.lineplot(x = 'Date', y = 'Confirmed', hue='Country', data = temp_df)

plt.show()
```
