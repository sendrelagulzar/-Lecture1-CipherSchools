# Basic Plotting with matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(11)
y=x**2
plt.plot(x,y)
print(plt.show())
plt.plot(y,x)
print(plt.show())
```

```python
plt.plot(x, y, 'b+') #b+ blue color + symbol
plt.show()
```

```python
# Labeling

plt.plot(x, y, 'ro', x, y, 'k-') # ro red dot    k- black line

plt.xlabel('Number(x)')
plt.ylabel('Square(y)')
plt.title('Scatter Plot - x & y')

plt.xlim([0, 11])
plt.ylim([0, 121])

plt.show()
```

```python
plt.plot([1, 3, 5, 7, 9], [1, 8, 3, 5, 1])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot - x & y')

plt.xlim([0, 10])
plt.ylim([0, 10])

plt.show()
```

```python
# Labeling the plot

heights = np.array([160, 180, 146, 162, 184, 180])

weights = np.array([50, 78, 45, 51, 80, 60])

plt.plot(heights, weights, 'ko')

plt.xlabel('Heights (cm)')
plt.ylabel('Weights (kg)')
plt.title('Scatter Plot - Height & Weight')

plt.xlim([140, 200])
plt.ylim([40, 90])

plt.show()
```

```python
# Multiple plots in a single figure

x = np.arange(1, 21)
y_1 = np.log(x)
y_2 = np.exp(x)

# Initiating a figure
plt.figure()

# Creating a plot with 1 row and 2 cols: subplot(row,col,index)

plt.subplot(1, 2, 1)
plt.title('y = log(x)')
plt.plot(x, y_1)

plt.subplot(1, 2, 2)
plt.title('y = exp(x)')
plt.plot(x, y_2)

plt.show()
```

Lets now look at some common plots using matplotlib
1. Box Plot (Univariate analysis)
2. Scatter Plot (Bivariate analysis)
3. Histograms (Univariate analysis)

### Box Plot

1. Very useful to detect outliers.
2. Shows 25th, 50th and 75th percentiles.

Percentile basically says where a point lies in a sorted list.<br />

Median-> 50th percentile<br />
Quantiles-> 25th, 50th, 75th, 100th percentile<br /><br />

75th percentile-> 75 percent of all the observations are less than 75th percentile.

Eg: CAT Rank

```python
df.columns
```

```python
plt.boxplot(df['Temperature'])
plt.title('Temprature boxplot')
plt.show()
```

```python
plt.boxplot(df['WindDirDegrees'])
plt.show()
```

### Histograms
1. Useful for visuallising distribution of 1 variable

```python
plt.hist(df['WindDirDegrees'])
plt.show()
```

### Scatter Plot

shows relation b/w 2 variable (karls person corretation

```python
plt.scatter(df['Temperature'], df['DewPoint'])
plt.xlabel('Temperature')
plt.ylabel('DewPoint')
plt.title('Scatter Plot - Temperature & DewPoint')
plt.show()
```

# Plotting with Seaborn

## 1 Univariate Analysis (for Numerical Variables)
i.e Analyzing one variable at a time.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('random.csv')

print(df.shape)

df.head()
```

```python
sns.displot(df['y'], kde=True)

# by default-> sns.distplot(x, hist=True, kde=True)
```

```python
#show probability distribution function (PDF) of y
sns.displot(df['y'], kde=True,stat = 'density')
```

### 1.1 Histogram

<b>REFERENCE: </b>https://statistics.laerd.com/statistical-guides/understanding-histograms.php

```python
sns.displot(df['y'], kde=False, rug=True)

# This is basically a frequency plot
# Try 'rug=False'
```

### 1.2 Kernel Density Estimation (1D Dimensionality plot)

<b>REFERENCE: </b>http://www.mvstat.net/tduong/research/seminars/seminar-2001-05/

```python
sns.displot(df['y'],kind='kde',rug=True)
```
